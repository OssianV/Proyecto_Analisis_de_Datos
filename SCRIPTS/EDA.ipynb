# EDA of the sqlite database

---

First we need to import the libraries to use

```python
import sqlite3
import pandas as pd
```

We now need to create a connection with the sqlite database

```python
connection = sqlite3.connect("../INPUTS/database.sqlite")

# We also print all the table names in the databse

cursor = connection.cursor()    # Create a cursor object
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()

for table in tables:
    print(table[0])
```

We start by performing a simple EDA of the table Match

```python
df_match = pd.read_sql_query("SELECT * FROM Match", connection)

display(df_match.columns)
```
